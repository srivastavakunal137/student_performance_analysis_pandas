# Module 3: Data Cleaning

In this module, the dataset is cleaned by removing duplicate records, handling missing values, validating the data, and preparing it for further analysis.

```python
# Number of rows before removing duplicates
print("Rows before removing duplicates: ", len(df))

#Remove duplicate rows
df = df.drop_duplicates()

# Number of rows after removing duplicates
print("Rows after removing duplicates: ", len(df))
```

```python
# Fill missing values

df = df.fillna({"StudyHours": df["StudyHours"].mean(),
               "Attendance": df["Attendance"].mean(),
               "Marks": df["Marks"].mean(),
               "Name":"Unknown"})
```

```python
# Validating Attendance

df = df[(df["Attendance"] >= 0) & (df["Attendance"] <= 100)]

print(df["Attendance"].min())
print(df["Attendance"].max())
```

```python
# Validating StudyHours 

df = df[df["StudyHours"] >= 0]
print(df["StudyHours"].min())
```

```python
# Validating Marks 

df = df[(df["Marks"] >= 0) & (df["Marks"] <= 100)]

print(df["Marks"].min())
print(df["Marks"].max())
```

```python
df.to_csv("../output/cleaned_data.csv", index = False)
print("Cleaned dataset saved successfully!")
```

### Observation

- Duplicate records were removed.
- Missing values were handled.
- Attendance values were validated.
- Study hours were validated.
- Marks were validated.
- Cleaned dataset was exported successfully.

# Module 4: Data Transformation

In this module, new features are created from the existing dataset. These include Grade, Result, and Performance Score, which help in evaluating each student's overall performance.

```python
def calculate_grade(marks):
    if marks >= 90:
        return "A"
    elif marks >= 75:
        return "B"
    elif marks >= 60:
        return "C"
    elif marks >= 40:
        return "D"
    else:
        return "F"
df["Grade"] = df["Marks"].apply(calculate_grade)
```

```python
df[["Name", "Marks", "Grade"]].head()
```

```python
df["Result"] = df["Marks"].apply( lambda x:"Pass" if x >= 40 else "Fail")
```

```python
df[["Name", "Marks", "Result"]].head()
```

```python
#  Performance Score calculation

df["Performance Score"] = (df["Marks"] * 0.60 + df["Attendance"] * 0.25 + df["StudyHours"] * 2) 
```

```python
df.head()
```

```python
df.to_csv("../output/cleaned_data.csv", index = False)
print("Dataset updated successfully!")
```

### Observation

- Grade column was created based on marks.
- Result column identifies Pass or Fail.
- Performance Score combines Marks, Attendance, and Study Hours into a single score.
- Updated dataset was saved successfully.

# Module 6: Data Analysis

## Objective

The objective of this module is to analyze the student dataset by calculating important performance metrics such as average marks, highest and lowest marks, attendance, study hours, pass percentage, fail percentage, and grade distribution.

```python
# Calculate analysis metrics

average_marks = df["Marks"].mean()
highest_marks = df["Marks"].max()
lowest_marks = df["Marks"].min()

average_attendance = df["Attendance"].mean()
average_study_hours = df["StudyHours"].mean()

pass_percentage = (df["Result"] == "Pass").mean() * 100
fail_percentage = (df["Result"] == "Fail").mean() * 100

analysis_summary = pd.DataFrame({
    "Metric": [
        "Average Marks",
        "Highest Marks",
        "Lowest Marks",
        "Average Attendance (%)",
        "Average StudyHours",
        "Pass Percentage(%)",
        "Fail Percentage(%)"
    ],
    "Value": [
        round(average_marks, 2),
        highest_marks,
        lowest_marks,
        round(average_attendance, 2),
        round(average_study_hours, 2),
        round(pass_percentage, 2),
        round(fail_percentage, 2)
    ]
})

analysis_summary
```

```python
grade_distribution = (
    df["Grade"]
    .value_counts()
    .reset_index()
)

grade_distribution.columns = ["Grade", "Number of Students"]

grade_distribution
```

```python
# Highest Scoring Student

highest_student = df.nlargest(
    1,
    "Marks"
)[["ID", "Name", "Marks", "Grade"]]

highest_student
```

```python
# Lowest Scoring Student

lowest_student = df.nsmallest(
    1,
    "Marks"
)[["ID", "Name", "Marks", "Grade"]]

lowest_student
```

```python
top10 = df.nlargest(
    10,
    "Performance Score"
)[["ID", "Name", "Marks",
   "Attendance",
   "StudyHours",
   "Performance Score"]]

top10
```

### Observation

- Calculated average, highest, and lowest marks.
- Computed average attendance and study hours.
- Calculated pass and fail percentages.
- Analyzed the distribution of grades.
- Identified the highest and lowest scoring students.
- Displayed the top 10 students based on Performance Score.

# Module 8: Grouping

## Objective

The objective of this module is to group students based on their grades and calculate important statistics such as average marks, number of students, and average attendance for each grade.

```python
average_marks_by_grade = (
    df.groupby("Grade")["Marks"]
      .mean()
      .reset_index()
)

average_marks_by_grade.columns = ["Grade", "Average Marks"]

average_marks_by_grade
```

```python
students_per_grade = (
    df.groupby("Grade")
      .size()
      .reset_index(name="Number of Students")
)

students_per_grade
```

```python
average_attendance_by_grade = (
    df.groupby("Grade")["Attendance"]
      .mean()
      .reset_index()
)

average_attendance_by_grade.columns = [
    "Grade",
    "Average Attendance"
]

average_attendance_by_grade
```

```python
grade_summary = (
    df.groupby("Grade")
      .agg(
          Average_Marks=("Marks", "mean"),
          Average_Attendance=("Attendance", "mean"),
          Student_Count=("Grade", "count")
      )
      .reset_index()
)

grade_summary
```

```python
grade_summary.to_csv("../output/grade_summary.csv", index=False)

print("Grade summary saved successfully!")
```

### Observation

- Students were grouped according to their grades.
- Average marks were calculated for each grade.
- Average attendance was calculated for each grade.
- Number of students in each grade was determined.
- Grade summary was exported successfully.

# Module 9: Statistical Analysis

## Objective

The objective of this module is to perform statistical analysis on the student dataset by calculating measures such as mean, median, mode, standard deviation, variance, and correlation between numerical attributes.

# Module 10: Report Generation

## Objective

The objective of this module is to generate a final summary report containing key statistics of the student dataset. The report includes total students, pass/fail counts, highest and lowest marks, average marks, average attendance, and grade-wise distribution.

```python
# Total Students
total_students = len(df)

# Passed Students
passed_students = len(df[df["Result"] == "Pass"])

# Failed Students
failed_students = len(df[df["Result"] == "Fail"])

# Highest Marks
highest_marks = df["Marks"].max()

# Lowest Marks
lowest_marks = df["Marks"].min()

# Average Marks
average_marks = round(df["Marks"].mean(), 2)

# Average Attendance
average_attendance = round(df["Attendance"].mean(), 2)

print("Calculations completed successfully!")
```

```python
grade_distribution = df["Grade"].value_counts()

grade_distribution
```

```python
report = pd.DataFrame({
    "Metric": [
        "Total Students",
        "Passed Students",
        "Failed Students",
        "Highest Marks",
        "Lowest Marks",
        "Average Marks",
        "Average Attendance (%)"
    ],
    "Value": [
        total_students,
        passed_students,
        failed_students,
        highest_marks,
        lowest_marks,
        average_marks,
        average_attendance
    ]
})

report
```

```python
for grade, count in grade_distribution.items():
    report.loc[len(report)] = [f"Grade {grade}", count]

report
```

```python
report.to_csv("../output/report.csv", index=False)

print("Report generated successfully!")
```
